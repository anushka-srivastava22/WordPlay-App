```python
# 1st Game

# !pip install requests
# !pip install beautifulsoup4

import requests #making https requests
from bs4 import BeautifulSoup
import csv
import random

class Pick_a_meaning:
    
    def question(self):
        self.words = []
        self.user_ans={}
        self.meanings = []
        self.word_meaning = {}
        
        #Open CSV File
        with open('final_EnglishGame_dataset.csv') as csv_file:
            csv_reader = csv.reader(csv_file, delimiter=',')
            for row in csv_reader:
              self.words.append(row[0])
        
        #Shuffling words
        random.shuffle(self.words)
        self.words = self.words[:5]
    #         meanings = []
    #         word_meaning = {}
    #     words = ["cat","bat","mat","sit","rat"]
        
        #Web Scrapping 
        for word in self.words:
          web = requests.get('https://www.lexico.com/en/definition/'+word)
          # print(word,":")
          data = web.content
          soup = BeautifulSoup(data,features="html.parser")
          # print(soup)
          tag = soup.find_all("span","ind")
          # print(tag)
          
          #Finding meaning and appending them 
          self.meaning_found = False
          for i in tag:
            self.meaning_found = True
            text = i.text
            self.meanings.append(text)
            self.word_meaning[word]=text
            break

          if(self.meaning_found==False):
            text = "Meaning Not Found"
            self.meanings.append(text)
            self.word_meaning[word]=text

    #         user_ans={}
        random.shuffle(self.words)
        random.shuffle(self.meanings)

#         print(self.words)
#         print(self.meanings)
#         print(self.word_meaning)
#         print(self.user_ans)
        
        return ("Match the given words with there meaning - \n\n")
    
    def word_meaning_display(self):

#         words = []
#         print(f'{"Words":{20}} {" ":{5}} Meanings(Options)')
        
        display = "\n"
        display+="Words - \n"
        
        #Showing word meaning to the user
        for i in range(len(self.words)):
            display+=str(chr(65+i))+") "+self.words[i]+"\n"
#             print(f'{str(chr(65+i))+") "+words[i]:{20}} {"-":{5}} {str(i+1)+") "+meanings[i]}')
        
        display+="\n\n"
        display+="Meanings - \n"
        
        
        for i in range(len(self.words)):
            display+=str(i+1)+") "+self.meanings[i]+"\n"
#             print(f'{str(chr(65+i))+") "+words[i]:{20}} {"-":{5}} {str(i+1)+") "+meanings[i]}')
        
        display+="\n\n"
        
        return str(display)
    
    # Showing words to get answers
    def answers(self):

        #print("\nYour answers here - ")
        
        ques_display = []
        for i in range(len(self.words)):
            temp_ques = str(chr(65+i))+") "+self.words[i]+" - "
            ques_display.append(temp_ques)
            #print(f'{str(chr(65+i))+") "+words[i]:{20}} {"-":{5}}',end=" ")
            #ans = int(input())

        return ques_display
    
    
    def score_eval(self, answer_list):
        
        #Access answer list and store it in our users answer dictionary
        idx = 0
        for ans in answer_list:
            ans = int(ans)
            self.user_ans[self.words[idx]] = self.meanings[ans-1]
            idx+=1
        
        #Calculating score
        score = 0
        for i in list(self.word_meaning.keys()):
            if(self.user_ans[i]==self.word_meaning[i]):
                score+=1
            
        #print(str("Score: "+str(score)+"/"+str(len(self.user_ans))))
        return str(str(score)+"/"+str(len(self.user_ans)))


        # print(words)
        # print(meanings)
        # print(word_meaning)
        # print(user_ans)
        #print("\nScore: ",score,"/",len(user_ans))
        
    #Showing correct answers
    def correct_answers(self):
        print("\nCorrect answers are - ")
        for i in list(self.word_meaning.keys()):
          print(i,"â–º",self.word_meaning[i],"\n")
        
    

# p = Pick_a_meaning()
# score = p.meaning_picker()
# print(score)
```
